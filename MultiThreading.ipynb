```python
from threading import *
```

```python
print(current_thread().getName())
```

```python
def myclass():
    for i in range(100):
        print(i,current_thread().getName())
        
t=Thread(target=myclass,name='Childthread')
t.start()

for i in range(10):
    print('Mainthread')
```

```python
class myclass(Thread):
    def run(self):
        for i in range(10):
            print('Child Thread',current_thread().name)
            print(i,current_thread().ident)
            
obj=myclass()
t=Thread(target=obj.run)
t.start()

for i in range(10):
    print(i,current_thread().getName())
    print(i,current_thread().ident)
    
print(active_count())
print(t.isAlive())
```
